```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn import cross_validation, linear_model, preprocessing
from sklearn.metrics import precision_recall_curve, precision_score, recall_score
from __future__ import division
%matplotlib inline
```

### Read non-crowdsourced data :

```python
raw_data = pd.read_csv('./learndata-latest.csv')
raw_data.head()
```

### Read crowdsourced data : 

```python
raw_data2 = pd.read_csv('./crowd_sourced_learn020916.csv')
raw_data2.head()
```

### Process crowdsourced data : 

```python
X2 = raw_data2

X2['same_curbtype'] = (X2['curbtype_i'] == X2['curbtype_j']).astype(int)
X2['same_surftype'] = (X2['surftype_i'] == X2['surftype_j']).astype(int)
X2['same_block'] = (X2['bid_i'] == X2['bid_j']).astype(int)
X2['same_side'] = (X2['side_i'] == X2['side_j']).astype(int)

# TODO: There's probably more features we can generate with these two lengths
X2['length_diff'] = abs(X2['length_i'] - X2['length_j'])
X2['sw_width_diff'] = abs(X2['sw_width_i'] - X2['sw_width_j'])

y2 = X2['connected']
X2 = X2.drop('connected', 1)
X2 = X2.drop('curbtype_i', 1)
X2 = X2.drop('curbtype_j', 1)
X2 = X2.drop('surftype_i', 1)
X2 = X2.drop('surftype_j', 1)

X2['near_angle'] = abs(X2['near_angle'])

del X2['bid_i']  # This has NaN and will cause errors in sklearn
del X2['bid_j']

del X2['side_i']
del X2['side_j']
del X2['length_i']
del X2['length_j']
del X2['sw_width_i']
del X2['sw_width_j']

# Scaling appropriate features by subtracting mean and scaling to unit variance
scaler = preprocessing.StandardScaler()

X2['near_distance'] = scaler.fit_transform(X2['near_distance'])
X2['length_diff'] = scaler.fit_transform(X2['length_diff'])
X2['sw_width_diff'] = scaler.fit_transform(X2['sw_width_diff'])

X2.head()
```

### Process non-crowdsourced data : 

```python
label_encoder = preprocessing.LabelEncoder()

near_line = raw_data['near_line']  # Save for later

del raw_data['near_line']
del raw_data['id_i']
del raw_data['id_j']

raw_data['curbtype_i'] = label_encoder.fit_transform(raw_data['curbtype_i'])
raw_data['curbtype_j'] = label_encoder.fit_transform(raw_data['curbtype_j'])
raw_data['side_i'] = label_encoder.fit_transform(raw_data['side_i'])
raw_data['side_j'] = label_encoder.fit_transform(raw_data['side_j'])
raw_data['surftype_i'] = label_encoder.fit_transform(raw_data['surftype_i'])
raw_data['surftype_j'] = label_encoder.fit_transform(raw_data['surftype_j'])
raw_data['intersects'] = label_encoder.fit_transform(raw_data['intersects'])


X = raw_data

X['same_curbtype'] = (X['curbtype_i'] == X['curbtype_j']).astype(int)
X['same_surftype'] = (X['surftype_i'] == X['surftype_j']).astype(int)
X['same_block'] = (X['bid_i'] == X['bid_j']).astype(int)
X['same_side'] = (X['side_i'] == X['side_j']).astype(int)

# TODO: There's probably more features we can generate with these two lengths
X['length_diff'] = abs(X['length_i'] - X['length_j'])
X['sw_width_diff'] = abs(X['sw_width_i'] - X['sw_width_j'])

y = X['connected']
X = X.drop('connected', 1)
X = X.drop('curbtype_i', 1)
X = X.drop('curbtype_j', 1)
X = X.drop('surftype_i', 1)
X = X.drop('surftype_j', 1)

X['near_angle'] = abs(X['near_angle'])

del X['bid_i']  # This has NaN and will cause errors in sklearn
del X['bid_j']

del X['side_i']
del X['side_j']
del X['length_i']
del X['length_j']
del X['sw_width_i']
del X['sw_width_j']

# Scaling appropriate features by subtracting mean and scaling to unit variance
scaler = preprocessing.StandardScaler()

X['near_distance'] = scaler.fit_transform(X['near_distance'])
X['length_diff'] = scaler.fit_transform(X['length_diff'])
X['sw_width_diff'] = scaler.fit_transform(X['sw_width_diff'])

X.head()
```

### Generate training and test splits for both data sets : 

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.9, random_state=6883)
X2_train, X2_test, y2_train, y2_test = cross_validation.train_test_split(X2, y2, test_size=0.5, random_state=6883)
```

### Training Decision Tree on Crowd and testing on non-crowd : 

```python
#learning a decision tree
from sklearn import tree
DT2 = tree.DecisionTreeClassifier()
DT2 = DT2.fit(X2_train, y2_train)
print(DT2.score(X_test, y_test))
Dt2_pred = DT2.predict(X_test)
print precision_score(y_test, Dt2_pred)
print recall_score(y_test, Dt2_pred)
```

### Training Decision Tree on non-crowd and testing on crowd : 

```python
#learning a decision tree
from sklearn import tree
DT = tree.DecisionTreeClassifier()
DT = DT2.fit(X_train, y_train)
print(DT.score(X2_test, y2_test))
Dt_pred = DT.predict(X2_test)
print precision_score(y2_test, Dt_pred)
print recall_score(y2_test, Dt_pred)
```

### Training Logistic Regression (Polynomial) on non-Crowd and testing on crowd : 

```python
poly = preprocessing.PolynomialFeatures(2)
Xp2 = poly.fit_transform(X2)
Xp = poly.fit_transform(X)
Xp_train, Xp_test, yp_train, yp_test = cross_validation.train_test_split(Xp, y, test_size=0.9, random_state=6883)
Xp2_train, Xp2_test, yp2_train, yp2_test = cross_validation.train_test_split(Xp2, y2, test_size=0.5, random_state=6883)
```

```python
logisticp = linear_model.LogisticRegression(penalty='l1', C=1e-2)
logisticp.fit(Xp_train, yp_train)
print 'Fit score: {}'.format(logisticp.score(Xp2_test, yp2_test))
y_predp = logisticp.predict(Xp2_test)
y_probp = logisticp.predict_proba(Xp2_test)[:,1]
precisionp, recallp, thresholdsp = precision_recall_curve(y2_test, y_probp)
print "Precision score (what fraction of predicted connections are true): {}".format(precision_score(y2_test, y_predp))
print "Recall score (what fraction of connections were found?): {}".format(recall_score(y2_test, y_predp))

plt.plot(recallp, precisionp)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```

```python
logisticp2 = linear_model.LogisticRegression(penalty='l1', C=1e-2)
logisticp2.fit(Xp2_train, yp2_train)
print 'Fit score: {}'.format(logisticp2.score(Xp_test, yp_test))
y2_predp = logisticp2.predict(Xp_test)
y2_probp = logisticp.predict_proba(Xp_test)[:,1]
precisionp2, recallp2, thresholdsp2 = precision_recall_curve(y_test, y2_probp)
print "Precision score (what fraction of predicted connections are true): {}".format(precision_score(y_test, y2_predp))
print "Recall score (what fraction of connections were found?): {}".format(recall_score(y_test, y2_predp))

plt.plot(recallp2, precisionp2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```
